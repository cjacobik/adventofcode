```python
turns=[]
with open('input.txt') as f:
    turns=[line.strip() for line in f.readlines()]
```

```python
#part 1
dial=50
# turns=['L68','L30','R48']
count0=0
for i in turns:
    lr=i[0]
    d=int(i[1:])
    newdial=(dial-d)%100 if lr=='L' else (dial+d)%100
    print(f'i {i} dial {dial} lr {lr} d {d} newdial {newdial}')
    count0+=1 if newdial==0 else 0
    dial=newdial
print(f'count0: {count0}')
```

```python
#part 2
dial=50
# turns=['L68','L30','R48']
count0=0
for i in turns:
    lr=i[0]
    d=int(i[1:])
    newdial=dial
    for ii in range(d):
        if lr=='L':
            newdial-=1
        elif lr=='R':
            newdial+=1
        newdial=newdial%100
        if newdial==0:
            count0+=1
    print(f'i {i} dial {dial} lr {lr} d {d} newdial {newdial}')
    dial=newdial
print(f'count0: {count0}')
```
