```python
#part 1
with open('day3.txt','r') as f:
    batteries=[i.strip() for i in f.readlines()]
vals=[]
for batt in batteries:
    a=max([int(i) for i in batt[:-1]])
    idx=batt.find(str(a))
    b=max([int(i) for i in batt[idx+1:]])
    vals.append(int(str(a)+str(b)))
sum(vals)
```

```python
#part 2
with open('input.txt','r') as f:
    batteries=[i.strip() for i in f.readlines()]

# b='1'*len(batteries[0])
# bits=[]
# for i in range(len(b)):
#     for j in range(i+1,len(b)):
#         for k in range(j+1,len(b)):
#             bits.append(b[:i]+'0'+b[i+1:j]+'0'+b[j+1:k]+'0'+b[k+1:])
# print(len(bits))

b='0'*len(batteries[0])
bits=[]
for i in range(len(b)):
    for j in range(i+1,len(b)):
        for k in range(j+1,len(b)):
            for i2 in range(k+1,len(b)):
                for j2 in range(i2+1,len(b)):
                    for k2 in range(j2+1,len(b)):
                        for i3 in range(k2+1,len(b)):
                            for j3 in range(i3+1,len(b)):
                                for k3 in range(j3+1,len(b)):
                                    for i4 in range(k3+1,len(b)):
                                        for j4 in range(i4+1,len(b)):
                                            for k4 in range(j4+1,len(b)):
                                                bits.append(b[:i]+'1'+b[i+1:j]+'1'+b[j+1:k]+'1'+
                                                            b[k+1:i2]+'1'+b[i2+1:j2]+'1'+b[j2+1:k2]+'1'+
                                                            b[k2+1:i3]+'1'+b[i3+1:j3]+'1'+b[j3+1:k3]+'1'+
                                                            b[k3+1:i4]+'1'+b[i4+1:j4]+'1'+b[j4+1:k4]+'1'+b[k4+1:]
                                                            )

maxjolts=[]
for batt in batteries:
    allbattbits=[]
    print(f'batt {batt}')
    for b in bits:
        battbits=[batt[i] for i in range(len(batt)) if b[i]=='1']
        allbattbits.append(int("".join(battbits)))
    maxjolts.append(max(allbattbits))
sum(maxjolts)
```
