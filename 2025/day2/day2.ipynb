```python
#part 1
with open('input.txt','r') as f:
    ss=f.readlines()

s=ss[0]

vals=[]
for r in s.split(','):
    a,b=r.split('-')
    for i in range(int(a),int(b)+1):
        l=len(str(i))
        if l%2==0:
            stri=str(i)
            if stri[:int(l/2)]==stri[int(l/2):]:
                vals.append(i)
sum(vals)
```

```python
#part 2
with open('input.txt','r') as f:
    ss=f.readlines()

s=ss[0]

import re
vals=[]
for r in s.split(','):
    a,b=r.split('-')
    for i in range(int(a),int(b)+1):
        l=len(str(i))
        stri=str(i)
        found=False
        for ii in range(1,l):
            if not found:
                matches=re.findall(fr'(?:{stri[:ii]})+',stri)
                for m in matches:
                    if m==stri:
                        vals.append(i)
                        found=True
sum(vals)
```
