```python
#part 1
with open('input.txt','r') as f:
    pts=[i.strip() for i in f.readlines()]

d={}
maxi=0
maxj=0
for i in range(len(pts)):
    val=pts[i]
    maxi=i if i>maxi else maxi
    for j in range(len(val)):
        if i not in d.keys():
            d[i]={}
        d[i][j]=val[j]
        maxj=j if j>maxj else maxj

def countroll(d,i,j):
    try:
        if d[i][j]=='@':
            return 1
        else:
            return 0
    except:
        return 0

totalcount=0
for i in range(maxi+1):
    for j in range(maxj+1):
        if d[i][j]=='@':
            count=0
            count+=countroll(d,i-1,j-1)
            count+=countroll(d,i,j-1)
            count+=countroll(d,i+1,j-1)
            count+=countroll(d,i-1,j)
            count+=countroll(d,i+1,j)
            count+=countroll(d,i-1,j+1)
            count+=countroll(d,i,j+1)
            count+=countroll(d,i+1,j+1)
            if count<4:
                totalcount+=1

totalcount
```
