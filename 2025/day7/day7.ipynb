```python
with open('input.txt','r') as f:
    inputs=[i.strip() for i in f.readlines()]

beams=[]
countsplitters=0
for i in inputs:
    if 'S' in i:
        beams.append(i.find('S'))
    elif len(beams)>0:
        js=[ii for ii,char in enumerate(i) if char=='^']
        for j in js:
            if j in beams:
                countsplitters+=1
                beams.remove(j)
                beams.append(j-1)
                beams.append(j+1)
                beams=list(set(beams))
    print(f'i {i} beams {beams}')

countsplitters
```

```python
with open('day7.txt','r') as f:
    inputs=[i.strip().split() for i in f.readlines()]
print(inputs)

manifold = open("day7.txt", "r").read()

manifold = [list(row.strip()) for row in manifold.split()]
beam_traces = [[0 for entry in row] for row in manifold]

print(manifold)
print(beam_traces)

def trace_beam(x, y, manifold, beam_traces):
    if y >= len(manifold):
        print("beam exited")
        return 1
    if beam_traces[y][x] != 0:
        print(f'x {x} y {y} beam_traces {beam_traces}')
        return beam_traces[y][x]
    if manifold[y][x] == "^":
        print(f'x {x} y {y} beam_traces {beam_traces}')
        beam_traces[y][x] = trace_beam(x - 1, y + 1, manifold, beam_traces) + trace_beam(x + 1, y + 1, manifold, beam_traces)
    else:
        print(f'x {x} y {y} beam_traces {beam_traces}')
        beam_traces[y][x] = trace_beam(x, y + 1, manifold, beam_traces)
    print(f'x {x} y {y} beam_traces {beam_traces}')
    return beam_traces[y][x]


timelines = 0

for i in range(len(manifold[0])):
    if manifold[0][i] == "S":
        timelines = trace_beam(i, 0, manifold, beam_traces)
print(timelines)



# beams=[]
# countsplitters=0
# for i in inputs:
#     if 'S' in i:
#         beams.append(i.find('S'))
#     elif len(beams)>0:
#         js=[ii for ii,char in enumerate(i) if char=='^']
#         for j in js:
#             if j in beams:
#                 countsplitters+=1
#                 # beams.remove(j)
#                 beams.append(j-1)
#                 beams.append(j+1)
#                 # beams=list(set(beams))
#     print(f'i {i} beams {beams}')

# len(beams)
```
