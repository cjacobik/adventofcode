```python
#part 1
import re,pprint
with open('input.txt','r') as f:
    inputs=[re.sub(r'\s+',' ',i.strip()) for i in f.readlines()]

cols={}
for i in inputs:
    vals=i.split(' ')
    for j in range(len(vals)):
        if j not in cols.keys():
            cols[j]=[]
        cols[j].append(vals[j])

finaltotal=0
for j,vals in cols.items():
    v=vals.pop()
    if v=='*':
        total=1
        for vv in vals:
            total*=int(vv)
    if v=='+':
        total=0
        for vv in vals:
            total+=int(vv)
    finaltotal+=total

print(finaltotal)
# pprint.pprint(cols)
```
